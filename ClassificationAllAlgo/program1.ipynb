## problem statement
- predict the customer's segment based on wine they purchased
- customer_segment will be the dependent variable

## import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## load the dataset

```python
df = pd.read_csv("C:\\Users\\sainya.vishwakarma\\Python\\ClassificationAllAlgo\\Wine.csv")
df
```

### exploratory data analysis

### find the correlation analysis

```python
# drop ash,magnesium and color_intensity
df.drop(['Ash','Magnesium','Color_Intensity'],axis=1,inplace=True)
```

```python
# check if the dataset is balanced
df['Customer_Segment'].value_counts()
```

### split the dataset into x and y

```python
x = df.drop('Customer_Segment',axis=1)
y=df['Customer_Segment']
```

```python
# change y
y_new=y-1
```

```python
y_new
```

### split the data into train and test dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y_new,train_size=0.7,random_state=42)
```

### train the models

```python
from sklearn.linear_model import LogisticRegressionCV

def create_lr_model():
    model = LogisticRegressionCV(max_iter=5000)
    model.fit(x_train,y_train) 
    return model
```

```python
from sklearn.naive_bayes import GaussianNB

def create_nb_model():
    model = GaussianNB()
    model.fit(x_train,y_train) 
    return model
```

```python
from sklearn.svm import SVC

def create_svc_model():
    model = SVC(C = 10, kernel='rbf')
    model.fit(x_train,y_train) 
    return model
```

```python
from sklearn.neighbors import KNeighborsClassifier

def create_knn_model():
    model = KNeighborsClassifier(n_neighbors=3)
    model.fit(x_train,y_train) 
    return model
```

```python
from sklearn.tree import DecisionTreeClassifier

def create_dt_model():
    model = DecisionTreeClassifier(max_depth=3)
    model.fit(x_train,y_train) 
    return model
```

```python
from sklearn.ensemble import RandomForestClassifier

def create_rf_model():
    model = RandomForestClassifier(n_estimators=3,max_depth=3)
    model.fit(x_train,y_train) 
    return model
```

```python
from sklearn.ensemble import AdaBoostClassifier

def create_ada_boost_model():
    model = AdaBoostClassifier()
    model.fit(x_train,y_train) 
    return model
```

```python
from sklearn.ensemble import GradientBoostingClassifier

def create_gb_boost_model():
    model = GradientBoostingClassifier(learning_rate=0.1,max_depth=4,n_estimators=10,verbose=1)
    model.fit(x_train,y_train) 
    return model
```

```python
from xgboost import XGBClassifier

def create_xgb_model():
    model = XGBClassifier()
    model.fit(x_train,y_train) 
    return model
```

```python
from lightgbm import LGBMClassifier

def create_lgbm_model():
    model = LGBMClassifier(learning_rate=0.1,n_estimators=10)
    model.fit(x_train,y_train) 
    return model
```

```python
from catboost import CatBoostClassifier

def create_catboost_model():
    model = CatBoostClassifier()
    model.fit(x_train,y_train) 
    return model
```

```python
# evaluate the models
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

def evaluate_model(model,name,x,y):
    y_pred = model.predict(x)
    accuracy = accuracy_score(y,y_pred)
    precision = precision_score(y,y_pred,average='micro')
    recall = recall_score(y,y_pred,average='micro')
    f1 = f1_score(y,y_pred,average='micro')
    return name,accuracy,precision,recall,f1
```

```python
# create the models

model_lr=create_lr_model()
model_nb=create_nb_model()
model_svm=create_svc_model()
model_knn=create_knn_model()
model_dt=create_dt_model()
model_rf=create_rf_model()
model_ada_boost=create_ada_boost_model()
model_gb=create_gb_boost_model()
model_xgb=create_xgb_model()
model_lgbm=create_lgbm_model()
model_catboot=create_catboost_model()
```

```python
# create list of all models

models=[
    {"name":"Logistic Regression","model":model_lr},
    {"name":"Naive Bayes","model":model_nb},
    {"name":"SVM","model":model_svm},
    {"name":"KNN","model":model_knn},
    {"name":"Decision Tree","model":model_dt},
    {"name":"Random Forest","model":model_rf},
    {"name":"Ada Boost","model":model_ada_boost},
    {"name":"Gradient Boost","model":model_gb},
    {"name":"XG Boost","model":model_xgb},
    {"name":"LGBM","model":model_lgbm},
    {"name":"Cat Boost","model":model_catboot},
]
```

```python
# evaluate all the models on train set

train_metrics=[]
for info in models:
    train_metrics.append(evaluate_model(info['model'],info['name'],x_train,y_train))
pd.DataFrame(train_metrics,columns=['Algorithm','Accuracy','Precision','Recall','F1 Score'])
```

```python
# evaluate all the models on test set

test_metrics=[]
for info in models:
    test_metrics.append(evaluate_model(info['model'],info['name'],x_test,y_test))
pd.DataFrame(test_metrics,columns=['Algorithm','Accuracy','Precision','Recall','F1 Score'])
```
