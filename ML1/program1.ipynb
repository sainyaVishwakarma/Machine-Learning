```python
plt.scatter(df["YearsExperience"],df["Salary"],label="Salary")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("Salary and Year of Experience of employee")
plt.legend()
plt.grid()
```

```python
# correlation analysis
# - correlation is the relationship between two variables
# - the correlation coefficient is used to meausure the relationship's direction and magnitude
# - the values stays between -1 to +1
# - interpretation
# - -1 : perfect negative relationshi[
# - 0: no relationship
# - +1:perfect positive relationship
df.corr()
```

```python
### split the data
```

```python
# split the data into X and Y
# X is independent variable, it is must that X must be two dimensional array or data frame
# - drop() will remove an entity
# - axis = 0, it will remove a row with the index
# - axis = 1, it will remove a column with the name 
x =df.drop('Salary',axis=1)

# it is mandatory that Y should be a one dimensional array or series object
y= df['Salary']
```

```python
x.shape
```

```python
df["YearsExperience"].ndim
```

```python
x.ndim
```

```python
y.ndim
```

```python
### train the model
```

```python
from sklearn.linear_model import LinearRegression

# create an object of LinearRegression
model = LinearRegression()

# train the model using x and y
model.fit(x,y)
```

```python
### result visualization
```

```python
# plot the original values
plt.scatter(df["YearsExperience"],df["Salary"],label="Salary")

# predict salaries for all given x
y_pred = model.predict(x)

# plot a regression line
plt.scatter(df["YearsExperience"],y_pred, color ="Red")
plt.plot(df["YearsExperience"],y_pred, color ="Red", label = "Regression Line")
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.title("Salary and Year of Experience of employees")
plt.legend()
plt.grid()
```

```python
### interpret the model
```

```python
# get the coefficients

model.coef_
```

```python
# get the intercept

model.intercept_
```
