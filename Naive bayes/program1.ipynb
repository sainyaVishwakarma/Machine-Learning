```python
### data preparation
```

```python
### split the data into x and y
```

```python
x = df.drop('test_result',axis=1)
y=df['test_result']
```

```python
### split the data into train and test dataset
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
### build the models
```

```python
from sklearn.linear_model import LogisticRegressionCV

def build_logistic_regression_model():
    # create the model
    model = LogisticRegressionCV()

    # fit the model
    model.fit(x_train,y_train)

    return model
```

```python
from sklearn.naive_bayes import GaussianNB 

def build_naive_bayes_model():
    # create the model
    model = GaussianNB()

    # fit the model
    model.fit(x_train,y_train)

    return model
```

```python
### evaluate models
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,roc_auc_score,roc_curve,log_loss

def evaluate_model(model_name,model,x,y):
    # predict x
    y_pred = model.predict(x)

    # calculate metrics
    accuracy = accuracy_score(y,y_pred)
    precision = precision_score(y,y_pred)
    recall = recall_score(y,y_pred)
    f1 = f1_score(y,y_pred)
    log= log_loss(y,y_pred)
    roc = roc_auc_score(y,y_pred)

    return model_name,accuracy,precision,recall,f1,log,roc
```

```python
# create all the  models
model_logistic_regression =  build_logistic_regression_model()
model_naive_bayes_regression = build_naive_bayes_model()

# create an array of models

models = [
    {"model_name": "Logistic Regression", "model":model_logistic_regression },
    {"model_name": "Naive Bayes", "model":model_naive_bayes_regression }
]
```

```python
# evaluate all the models on train set

evaluate_metrics=[]

for model_info in models:
    evaluate_metrics.append(evaluate_model(model_info['model_name'],model_info['model'],x_train,y_train))

pd.DataFrame(evaluate_metrics,columns=['model_name','accuracy','precision','recall','f1_score','log_loss','roc_auc_score'])
```

```python
# evaluate all the models on test set

evaluate_metrics=[]

for model_info in models:
    evaluate_metrics.append(evaluate_model(model_info['model_name'],model_info['model'],x_test,y_test))

pd.DataFrame(evaluate_metrics,columns=['model_name','accuracy','precision','recall','f1_score','log_loss','roc_auc_score'])
```
