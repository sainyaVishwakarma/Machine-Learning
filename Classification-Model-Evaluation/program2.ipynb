## problem statement
 - we have to predict if the customer will leave the bank
 - `Exited` will be the dependent variable


```python
### import required packages
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\sainya.vishwakarma\\Python\\Classification-Model-Evaluation\\Churn_Modelling.csv")
df
```

```python
### data preprocessing
```

```python
# since the rowNumber,CustomerId and Surname of no use while predicting exited value, we can remove it
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
### get the new list of columns
```

```python
df.columns
```

```python
### convert the categorical variables to numeric
```

```python
from sklearn.preprocessing import LabelEncoder

geography_encoder=LabelEncoder()
geography_encoder.fit(df['Geography'])
df['Geography'] = geography_encoder.transform(df['Geography'])

gender_encoder=LabelEncoder()
gender_encoder.fit(df['Gender'])
df['Gender'] = gender_encoder.transform(df['Gender'])
```

```python
# verify if all the data are numeric
df.info()
```

```python
# check if there are missing values
df.isna().sum()
```

```python
### perform corelation analysis
```

```python
df.corr()['Exited'].sort_values()
```

```python
### check if dataset is balanced
```

```python
# this is imbalanced dataset
df['Exited'].value_counts()
```

```python
### split the data into majority and minority class
df_majority = df[df['Exited']==0]
df_minority = df[df['Exited']==1]
```

```python
### oversampling
# - adding the random records to minority class to match with majority class
# - remove 5926 records to minority class to match with majority class 

from sklearn.utils import resample
df_oversampled = resample(df_minority,n_samples=7963) 
```

```python
df_oversampled.shape
```

```python
# merge the records to get the final dataset
df_new = pd.concat([df_oversampled,df_majority])
df_new.shape
```

```python
df_new['Exited'].value_counts()
```

```python
### split the data into x and y
```

```python
x = df_new.drop('Exited',axis=1)
y= df_new['Exited']
```

```python
### split the data int0 train and test set
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
### train the model
```

```python
from sklearn.linear_model import LogisticRegressionCV

# CV : Cross Validation
model = LogisticRegressionCV(max_iter=5000)

model.fit(x_train,y_train)
```

```python
### model evaluation
```

```python
# predict the x_test dataset
y_pred=model.predict(x_test)
y_pred
```

```python
### confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
tn,fp,fn,tp = cm.ravel()
print(f"tp: {tp}, tn: {tn}, fp: {fp},fn:{fn}")
```

```python
### accuracy
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)
print(f'accuracy= {accuracy * 100:.2f}%')

accuracy = (tp+tn)/(tp+tn+fp+fn)
print(f'accuracy= {accuracy * 100:.2f}%')
```

```python
### precision
```

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test,y_pred)
print(f'precision= {precision * 100:.2f}%')

precision = (tp)/(tp+fp)
print(f'precision= {precision * 100:.2f}%')
```

```python
### recall
```

```python
from sklearn.metrics import recall_score

recall = recall_score(y_test,y_pred)
print(f'recall= {recall * 100:.2f}%')

precision = (tp)/(tp+fn)
print(f'recall= {recall * 100:.2f}%')
```

```python
### f1_score
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test,y_pred) 
print(f'f1_score= {f1 * 100:.2f}%')

f2 = 2 * ((precision * recall)  / (precision + recall))
print(f'f1_score= {f2 * 100:.2f}%')
```

```python
### roc value

from sklearn.metrics import roc_auc_score

roc = roc_auc_score(y_test,y_pred) 
print(f'roc_auc_score= {roc:.2f}')
```

```python
### roc curve
```

```python
from sklearn.metrics import roc_curve
fpr,tpr,thresholds=roc_curve(y_test,y_pred)
plt.scatter(fpr,tpr,color='orange')
plt.plot(fpr,tpr,color='orange')
plt.plot([0,1],[0,1],'b--')
```
