### K-Means Clustering

## problem statement
- find the clusters/groups within the dataset based on the income and spending score

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\sainya.vishwakarma\\Python\\UnsupervisedLearning\\KMeansClustering\\Mall_Customers.csv")
df
```

```python
df.info()
```

```python
# since the clusters will be found based on income and spending score lets delete remaining columns
df.drop(['CustomerID','Gender','Age'],axis=1,inplace=True)
```

```python
df.rename({'Annual Income (k$)':'income','Spending Score (1-100)':'spending_score'}, axis=1,inplace=True)
```

### data visualization

```python
plt.scatter(df['income'],df['spending_score'])
plt.xlabel('income')
plt.ylabel('spending score')
```

### create clusters using Kmeans

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
kmeans.fit(df)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

### cluster visualization

```python
sns.scatterplot(df,x='income',y='spending_score',hue=kmeans.labels_)
```

```python
# add clusters
plt.scatter(df['income'][kmeans.labels_==0],df['spending_score'][kmeans.labels_==0],color='red',label='mid income mid spending score')
plt.scatter(df['income'][kmeans.labels_==1],df['spending_score'][kmeans.labels_==1],color='blue',label='low income low spending score')
plt.scatter(df['income'][kmeans.labels_==2],df['spending_score'][kmeans.labels_==2],color='green',label='high income low spending score')
plt.scatter(df['income'][kmeans.labels_==3],df['spending_score'][kmeans.labels_==3],color='black',label='high income high spending score')
plt.scatter(df['income'][kmeans.labels_==4],df['spending_score'][kmeans.labels_==4],color='magenta',label='low income low spending score')
plt.xlabel('income')
plt.ylabel('spending score')
plt.title("K-Means Clustering")
plt.legend()

for i in range(5):
    plt.scatter(kmeans.cluster_centers_[i][0],kmeans.cluster_centers_[i][1],color='brown',marker='^')
```

### evaluation of clusters

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(df,kmeans.labels_)
score
```

```python
kmeans6 = KMeans(n_clusters=6)
kmeans6.fit(df)
score = silhouette_score(df,kmeans6.labels_)
score
```

```python
kmeans4 = KMeans(n_clusters=4)
kmeans4.fit(df)
score = silhouette_score(df,kmeans4.labels_)
score
```
