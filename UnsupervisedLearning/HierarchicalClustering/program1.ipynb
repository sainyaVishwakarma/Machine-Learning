### hierarchical clustering

## problem statement
- find the clusters/groups within the dataset based on the income and spending score

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\sainya.vishwakarma\\Python\\UnsupervisedLearning\\HierarchicalClustering\\Mall_Customers.csv")
df
```

```python
df.info()
```

```python
# since the clusters will be found based on income and spending score lets delete remaining columns
df.drop(['CustomerID','Gender','Age'],axis=1,inplace=True)
```

```python
 df.rename({'Annual Income (k$)':'income','Spending Score (1-100)':'spending_score'}, axis=1,inplace=True)
```

### build the clusters

```python
from sklearn.cluster import AgglomerativeClustering
agglomerative_clustering=AgglomerativeClustering(n_clusters=5)
agglomerative_clustering.fit(df)
```

```python
agglomerative_clustering.labels_
```

### cluster visualization

```python
markers=['s','^','+','<','*']

for k in range(5):
    plt.scatter(df['income'][agglomerative_clustering.labels_==k],df['spending_score'][agglomerative_clustering.labels_==k],marker=markers[k])
    plt.xlabel('income')
plt.ylabel('spending score')
plt.title("Clustering")
```

### dendrogram visualization

```python
from scipy.cluster.hierarchy import linkage,dendrogram

linkage_single=linkage(df,method='single')
_=dendrogram(linkage_single)
```

```python
linkage_complete=linkage(df,method='complete')
_=dendrogram(linkage_complete)
```

```python
linkage_average=linkage(df,method='average')
_=dendrogram(linkage_average)
```
